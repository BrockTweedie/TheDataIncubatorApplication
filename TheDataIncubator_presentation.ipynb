# Virtual Art Appraiser App Status



## The pilot data analysis

- Scraped & cleaned text info for 10,000 19th-Century oil paintings from Sotheby's websites
    - Past 20 years of auctions
    - About 6 MB
    - 2400 distinct artists from 30 countries
    - Corrected for different currencies, inflation, sloppy/inconsistent data entry
    - Not exhaustive...there's a lot more data out there just in Sotheby's
- Train on 2000-2015 data
    - Baseline fixed guess model: global geometric mean ($35,400)
    - Categories & regression model: Construct categories according to frequently auctioned artist or nationality, run per-category 4-feature linear regressions vs painting size, aspect ratio, date painted, auction date (threshold: 15 paintings)
    - Test on 2016-2018 data
